```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score ,confusion_matrix
import seaborn as sns
```

```python
df=pd.read_csv("spambase_csv.csv")
x=df[df.drop('class',axis=1).columns]
y=df['class']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=39,test_size=0.3)
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
X_1=model.predict(x_test)
```

```python
print(accuracy_score(y_test,X_1))
```

```python
conf_matrix = confusion_matrix(y_test, X_1)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="magma")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
